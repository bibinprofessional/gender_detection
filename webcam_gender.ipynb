```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2
import os
import cvlib 
```

```python
gender_model = load_model('gender.model')
face_model = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
camera_video = cv2.VideoCapture(0)

while True:

    ret, frame = camera_video.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        face = frame[y:y+h, x:x+w]

        face = cv2.resize(face, (96,96))
        face = face/255.0
        face = np.expand_dims(face, axis=0)

        gender_prediction = gender_model.predict(face)
        gender = "Male" if gender_prediction[0][0] > 0.5 else "Female"

        cv2.putText(frame, f'Gender: {gender}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow('Face Detection and Gender Identification', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


camera_video.release()
cv2.destroyAllWindows()
```
